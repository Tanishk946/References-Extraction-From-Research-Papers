```python
from pdf2image import convert_from_path
import pytesseract
import os
import re
import nltk
from nltk.tag.stanford import StanfordNERTagger
nltk.internals.config_java("C:/Program Files/Java/jre1.8.0_281/bin/java.exe")
```

```python
def pic_to_text(pic_path):
    pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract'
    text=pytesseract.image_to_string(pic_path)
    file1 = open("pdf2text.txt", "a")
    file1.write(text)
    file1.close()
```

```python
def pdf_to_pic(pdf_path):
    pages = convert_from_path(pdf_path, 500,poppler_path=r'C:\poppler-0.68.0\bin')
    i=1
    for image in pages:
        image.save('output' + str(i) + '.jpg', 'JPEG')
        pic_to_text('output' + str(i) + '.jpg')
        os.remove('output' + str(i) + '.jpg')
        i=i+1
```

```python
pdf_to_pic('IEEE1.pdf')
```

```python
def filter_text(file_name):
    ref_line=0
    line_number=0
    read_obj=open(file_name,'r')
    for line in read_obj:
        line_number+=1
        if "References" in line or "REFERENCES" in line:
            ref_line=line_number
    read_obj.close()
    
    read_obj=open(file_name,'r')
    lines=0
    fp=open('references.txt','a+')
    for line in read_obj:
        lines+=1
        if lines>=ref_line:
            fp.write(line)
    fp.close()
    read_obj.close()
    os.remove('pdf2text.txt')
```

```python
filter_text('pdf2text.txt')
file1 = open("references.txt", "r")
print(file1.read())
file1.close()
```

# Part 2

```python
def author_removal(text):
    st = StanfordNERTagger('stanford-ner/english.all.3class.distsim.crf.ser.gz', 'stanford-ner/stanford-ner.jar')
    res=""
    for sent in nltk.sent_tokenize(text):
        tokens = nltk.tokenize.word_tokenize(sent)
        tags = st.tag(tokens)
        for tag in tags:
            if tag[1]=='PERSON' or tag[1]=='LOCATION' or tag[1]=='ORGANIZATION':
                continue
            else:
                res+=(tag[0]+" ")
    return res

ref=open("references.txt","r")
fp=open("author_remove.txt","a+")
lines=ref.readlines()
for line in lines:
    res=author_removal(line)
    print(res)
    fp.write(res+'\n')
ref.close()
fp.close()
os.remove("references.txt")
```

```python
rem=open("author_remove.txt","r")
fp1=open("punctuations_remove.txt","a+")
data=rem.readlines()
pattern='[0-9]'
for line in data:
    no_br_words=re.sub("\(.*?\)","()",line)
    no_punctuations= re.sub(r'[^\w\s]', '',no_br_words)
    no_digits=re.sub(pattern, '', no_punctuations)
    print(no_digits)
    fp1.write(no_digits+'\n')
rem.close()
fp1.close()
os.remove("author_remove.txt")
```

```python
pun=open("punctuations_remove.txt","r")
res=open("reference_nos.txt","a+")
count=0
l_no=1
no_of_lines=pun.readlines()
for i in range(len(no_of_lines)-1):
    if no_of_lines[i+1]!="\n":
        if count>=3:
            res.write(str(l_no)+"."+") ")
            res.write(no_of_lines[i+1]+"\n")
            l_no+=1
            count=0
        else:
            res.write(no_of_lines[i+1]+"\n")
        print(no_of_lines[i+1])
    else:
        count+=1
res.close()
pun.close()
os.remove("punctuations_remove.txt")
```
